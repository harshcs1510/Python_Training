```python
#dict comprehension

#Given values and keys create a dictionary using dict comprehension

k=['a','b','c']
v=[1,2,3]
d={k[i]:v[i] for i in range(len(k))}
#d={i:j for i,j in zip(k,v)}
d
```

```python
#Invert dictionary

dc={'a':1,'b':2,'c':3}
#d={i:j for j,i in dict.items(dc)}
d={dc[i]:i for i in dict.keys(dc)}
d
```

```python
#map(), filter(), reduce()

#map(func,iter)

def area(a):
    return 3.14*a**2

l=[2,3,4,5,6]
ls=list(map(area,l)) #passing def function
print(ls)
list(map(lambda a:3.14*a**2,l)) #passing lambda function
```

```python
#filter(func,data)

l=[2,3,4,5,6,7]

list(filter(lambda a:a%2==0,l))
```

```python
#reduce

from functools import reduce
data=[2,3,5,7,11,13,17]
multiply=lambda x,y:x*y
reduce(multiply,data)
```

```python
#Square each no. in list
l=[6,36,216]

s=lambda x:x**2
print(list(map(s,l)))

#Convet string list to uppercase
words=['bocchi','ikuyo','nijika','ryo']
uc=lambda w:w.upper()

print(list(map(uc,words)))

#Add two lists element wise

l1=[1,2,3]
l2=[4,5,6]

add=lambda i,j: i+j
print(list(map(add,l1,l2)))
```
